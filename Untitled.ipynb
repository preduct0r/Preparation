```python
def get_stat(file_name):
    data = pd.read_csv( file_name,skiprows=[0], header=None,names=["label"])
    t = data.reset_index()[data.label.shift()!=data.label].assign(f_name=file_name)
    t['to'] = t['index'].shift(-1)
    t['cnt']=  t['to']- t['index']
    
    t['size'] = t.cnt/30
    return t

get_stat(file_name)
```

```python
data = pd.concat([get_stat(i)
    for i in glob.glob(r"C:\Users\kotov-d\Documents\Aff_Beh\annotations\EXPR_Set\Training_Set\*")
])
```
